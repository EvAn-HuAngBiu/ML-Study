```python
import numpy as np
import pandas as pd

data = pd.read_table('testSet.txt', names=['val1', 'val2', 'label'], header=None)
data
```

```python
vals = data.iloc[:, :-1].values
labels = data.iloc[:, -1].values

from matplotlib import pyplot as plt
color_point = np.asarray([2 if i == 1 else 3 for i in labels], dtype=np.int)
plt.scatter(vals[:, 0], vals[:, 1], color_point * 15.0, color_point * 15.0)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def batch_gradient(x, y):
    y = y.reshape((-1, 1))
    x = np.c_[np.ones(x.shape[0]), x]
    m, n = x.shape
    alpha = 0.001
    max_cycle = 1500
    w = np.ones((n, 1))
    for k in range(max_cycle):
        h = sigmoid(np.dot(x, w))
        error = y - h
        w = w + alpha * np.dot(x.T,  error)
    return w

w = batch_gradient(vals, labels)
w
```

```python
plt.scatter(vals[:, 0], vals[:, 1], color_point * 15.0, color_point * 15.0)
tx = np.linspace(-4, 4)
ty = (-w[0] - w[1] * tx) / w[2]
plt.plot(tx, ty)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(max_iter=500)
classifier.fit(vals, labels)
n_w = classifier.coef_
n_b = classifier.intercept_

n_w, n_b
```

```python
plt.scatter(vals[:, 0], vals[:, 1], color_point * 15.0, color_point * 15.0)
tx = np.linspace(-4, 4)
ty = (-n_b - n_w[0][0] * tx) / n_w[0][1]
plt.plot(tx, ty)
plt.show()
```
