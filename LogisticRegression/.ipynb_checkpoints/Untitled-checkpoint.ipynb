```python
import pandas as pd
import numpy as np

test_path = 'horseColicTest.txt'
train_path = 'horseColicTraining.txt'

test_data = pd.read_table(test_path, header=None)
train_data = pd.read_table(train_path, header=None)

train_data
```

```python
testX = test_data.iloc[:, :-1].values
testY = test_data.iloc[:, -1].values
trainX = train_data.iloc[:, :-1].values
trainY = train_data.iloc[:, -1].values

testX
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter=500)
clf.fit(trainX, trainY)
pred = clf.predict(testX)

pred
```

```python
from sklearn.metrics import classification_report

t = classification_report(testY, pred, target_names=['0', '1'])
print(t)
```
