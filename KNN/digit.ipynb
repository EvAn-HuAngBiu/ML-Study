# 利用KNN实现灰度数字图识别

#### 定义数据读取函数，将文本文件中的32x32的数据读取到一个1024维行向量中

```python
import numpy as np
import matplotlib.pyplot as plt

def read_file_data(file_name):
    img_lines = []
    with open(file_name, 'r') as f:
        cur_line = f.readline().strip()
        while cur_line:
            str_list = list(cur_line)
            img_lines.extend(list(map(int, str_list)))
            cur_line = f.readline().strip()
    return img_lines
```

#### 读取训练数据集

```python
import os

train_path = r'/Users/evan/PythonProject/StatisticalLearning/CH3/digits/trainingDigits/'
train_data = []
train_label = []
for file_name in os.listdir(train_path):
    train_data.append(read_file_data(train_path + file_name))
    train_label.append(int(file_name[0]))
train_data = np.asarray(train_data)
train_label = np.asarray(train_label)

train_data.shape
```

#### 读取测试数据集

```python
test_path = r'/Users/evan/PythonProject/StatisticalLearning/CH3/digits/testDigits/'
test_data = []
test_label = []
for file_name in os.listdir(test_path):
    test_data.append(read_file_data(test_path + file_name))
    test_label.append(int(file_name[0]))
test_data = np.asarray(test_data)
test_label = np.asarray(test_label)

test_data.shape
```

#### 绘制测试灰度图像

```python
def plot_data(img_vec: np.ndarray):
    img_vec = np.asarray(list(map(lambda x: 0 if x == 0 else 255, img_vec)))
    img_mat = img_vec.reshape((32, 32))
    plt.imshow(img_mat, cmap=plt.cm.gray)
    plt.show()
    
plot_data(train_data[0])
```

#### 利用sklearn训练模型

关于KNeighborsClassifier分类器的参数说明：
- 定义：sklearn.neighbors.KNeighborsClassifier(n_neighbors=5, weights=’uniform’, algorithm=’auto’, leaf_size=30, 
             p=2, metric=’minkowski’, metric_params=None, n_jobs=None, **kwargs)
- n_neighbors: 邻近节点数量，默认为5，实际就是KNN里的K
- weight: 节点权重，指每一个节点在计算距离时是否具有相同权重，或不同权重的计算方法
   - uniform: 相同权重
   - distance: 离簇中心越近的点权重越高
   - callable: 用户自定义函数
- algorithm: 算法选择
   - auto: 自动选择最合适的算法
   - ball_tree: 使用BallTree
   - kd_tree: 使用KDTree
   - brute: 使用Brute-Force查找
- leaf_size: 传递给BallTree或者KDTree构造树的大小
- p: 设置Minkowski距离的Power参数，当p=1时，等价于manhattan距离；当p=2等价于euclidean距离，当p>2时，就是Minkowski 距离
- metric: 设置计算距离的方法
- metric_paras: 传递给metric方法的参数
- n_jobs: 并发执行的job数量

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10, weights="uniform", algorithm="auto", leaf_size=30, p=2, metric="minkowski",
                          metric_params=None, n_jobs=1)
knn.fit(train_data, train_label)
prediction = knn.predict(np.asarray([test_data[0]]))
print("Prediction is %d, reality is %d" % (prediction, test_label[0]))
```

#### 随机抽取部分图片验证

```python
test_img_num = test_label.shape[0]
plt.figure(figsize=(10, 10))
for i in range(25):
    index = np.random.randint(0, test_img_num, 1)
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_data[index].reshape((32, 32)), cmap=plt.cm.gray)
    plt.xlabel(str(test_label[index]) + "-" + str(knn.predict(np.asarray(test_data[index]))))
plt.show()
```

#### 模型评估

```python
access_score = knn.score(test_data, test_label)
access_score
```

#### 选择最佳K值

```python
scores = []
for cur_k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=cur_k)
    knn.fit(train_data, train_label)
    cur_score = knn.score(test_data, test_label)
    scores.append(cur_score)
plt.xticks(range(1, 20))
plt.plot(range(1, 20), scores, 'c-')
plt.show()
```

由上折线图可以看出，当k取值在3时，系统整体的误差率最小

```python
best_knn = KNeighborsClassifier(n_neighbors=3)
best_knn.fit(train_data, train_label)
best_score = best_knn.score(test_data, test_label)
best_score
```
